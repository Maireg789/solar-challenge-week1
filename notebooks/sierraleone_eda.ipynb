```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import zscore

pd.set_option('display.max_columns', None)
%matplotlib inline
```

```python
df = pd.read_csv("data/sierraleone-bumbuna.csv")
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
numeric_cols = df.select_dtypes(include=np.number).columns
z_scores = np.abs(zscore(df[numeric_cols]))
outliers = (z_scores > 3).sum()
outliers
```

```python
df_clean = df.copy()
df_clean[numeric_cols] = df_clean[numeric_cols].fillna(df_clean[numeric_cols].median())
df_clean = df_clean[(z_scores < 3).all(axis=1)]
df_clean.to_csv("data/sierraleone-bumbuna_clean.csv", index=False)
```

```python
plt.figure(figsize=(12, 4))
plt.plot(df_clean['Timestamp'], df_clean['GHI'], label='GHI', color='orange')
plt.plot(df_clean['Timestamp'], df_clean['ModA'], label='ModA', color='blue')
plt.plot(df_clean['Timestamp'], df_clean['ModB'], label='ModB', color='green')
plt.legend()
plt.title("Solar Metrics Over Time")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 4))
plt.plot(df['ModA'], label='Original ModA', alpha=0.5, color='gray')
plt.plot(df_clean['ModA'], label='Cleaned ModA', alpha=0.8, color='blue')
plt.legend()
plt.title("ModA Before vs After Cleaning")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df_clean.corr(), annot=True, cmap='Spectral')
plt.title("Correlation Matrix")
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(df_clean['WS'], bins=30, kde=True, color='purple')
plt.title("Wind Speed Distribution")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(x='RH', y='Tamb', data=df_clean, color='teal')
plt.title("Humidity vs Temperature")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df_clean['Tamb'], df_clean['GHI'], s=df_clean['WS']*10, alpha=0.5, c='red')
plt.xlabel("Temperature")
plt.ylabel("GHI")
plt.title("GHI vs Tamb (Bubble size = WS)")
plt.show()
```
